```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('img/tony.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('img/rainb.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img3 = cv2.imread('img/briks.jpg')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

```

```python
plt.imshow(img3)
```

```python
hist_val = cv2.calcHist([img1],channels=[0],mask=None,histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_val)
```

```python
hist_val = cv2.calcHist([img2],channels=[0],mask=None,histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_val)
```

```python
hist_val = cv2.calcHist([img3],channels=[0],mask=None,histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_val)
```

```python
img = img3
```

```python
color = ('b','g','r')

for i, col in enumerate(color):
    hist = cv2.calcHist([img], [i], None,[256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])
    
plt.title("Histogram of Bricks")    
```
