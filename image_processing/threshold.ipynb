```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img/rainb.jpg', 0)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img, cmap='gray')
```

```python
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```
