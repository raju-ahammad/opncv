```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread('img/sud.jpg', 0)
```

```python
display_img(img)
```

```python
soblex = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(soblex)
```

```python
sobley = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
display_img(sobley)
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
```

```python
display_img(laplacian)
```

```python
blended = cv2.addWeighted(src1=soblex,alpha=0.5, src2=sobley, beta=0.5, gamma=0)
```

```python
display_img(blended)
```

```python
ret, th1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
karnel = np.ones((4,4), dtype=np.uint8)
```

```python
gradient = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,karnel)
```

```python
display_img(gradient)
```
