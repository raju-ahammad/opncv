```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(img, text="RAJU", org=(50,300), fontFace=font, fontScale=5, color=(255,255,255), thickness=10)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
karnel = np.ones((5,5), dtype=np.uint8)
```

```python
res = cv2.erode(img, karnel, iterations=2)
```

```python
display_img(res)
```
