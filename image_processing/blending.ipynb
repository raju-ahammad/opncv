```python
img1 = cv2.imread('img/tony.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('img/stop.jpeg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray, cmap='gray')
```

```python
mask = cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
mask.shape
```

```python
import numpy as np
```

```python
white_bg = np.full(img2.shape, 255, dtype=np.uint8)
```

```python
white_bg.shape
```

```python
bg = cv2.bitwise_or(white_bg, white_bg, mask = mask)
```

```python
plt.imshow(bg)
```

```python
fg = cv2.bitwise_or(img2, img2, mask = mask)
```

```python
plt.imshow(fg)
```
