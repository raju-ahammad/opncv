```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
img = cv2.imread('img/tony.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img1 = cv2.imread('img/beuty.jpeg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
img1 = cv2.resize(img1, (600,700))
plt.imshow(img1)
```
