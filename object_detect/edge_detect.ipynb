```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('img/dog.jpg')
```

```python
plt.imshow(img)
```

```python
edge = cv2.Canny(image=img,threshold1=127, threshold2=127)
```

```python
plt.imshow(edge)
```

```python
edge = cv2.Canny(image=img,threshold1=155, threshold2=100)
plt.imshow(edge)
```

```python
med = np.median(img)
```

```python
med
```

```python
lower = int(max(0, 0.7*med))
upper = int(min(0,0.3*med))
```

```python
edge = cv2.Canny(image=img,threshold1=lower, threshold2=upper)
plt.imshow(edge)
```

```python
blur_img = cv2.blur(img,ksize=(8,8))
```

```python
edge = cv2.Canny(image=blur_img,threshold1=lower, threshold2=upper+50)
plt.imshow(edge)
```
