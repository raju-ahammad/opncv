```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
coins = cv2.imread('img/coins.jpg')
```

```python
display(coins)
```

```python
#Water shed Algorithm

#1Median Blur
#2Gray Scale
#3Binary ThreshHold
#Find Conturs
```

```python
blur_coins = cv2.medianBlur(coins, 25)
```

```python
display(blur_coins)
```

```python
gray_coins = cv2.cvtColor(coins, cv2.COLOR_BGR2GRAY)
```

```python
display(gray_coins)
```

```python
ret,coin_thresh = cv2.threshold(gray_coins,150,255, cv2.THRESH_BINARY_INV)
```

```python
display(coin_thresh)
```

```python
img, conturs, hierarchy = cv2.findContours(coin_thresh.copy(), cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in range(len(conturs)):
    
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(coins,conturs,i,(255,0,0),10)
```
