```python
chess = cv2.imread('img/ches.jpg')
chess = cv2.cvtColor(chess, cv2.COLOR_BGR2RGB)
gray_chess = cv2.cvtColor(chess, cv2.COLOR_BGR2GRAY)
```

```python
gray = np.float32(gray_chess)
```

```python
dst = cv2.cornerHarris(src = gray, blockSize=2, ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
chess[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(chess)
```

```python
chess = cv2.imread('img/chess.jpeg')
chess = cv2.cvtColor(chess, cv2.COLOR_BGR2RGB)


real_chess = cv2.imread('img/ches.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)


gray_chess = cv2.cvtColor(chess, cv2.COLOR_BGR2GRAY)
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_chess,49,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(chess,(x,y),8,(255,0,0),-3)
```

```python
plt.imshow(chess)
```
