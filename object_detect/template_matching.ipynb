```python
# All the 6 methods for comparison in a list

methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR', 'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 
                'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    
    #Create a copy full image
    fullimg_copy = full_img.copy()
    
    method = eval(m)
    
    #Template matching
    res = cv2.matchTemplate(fullimg_copy, face, method)
    
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
    
    if method in[cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc #(x,y value)
    else:
        top_left = max_loc
        
    height,width,channels = face.shape    
    
    bottom_right = (top_left[0]+width, top_left[1]+height)
    
    cv2.rectangle(fullimg_copy, top_left, bottom_right, (255,0,0), 10)
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title("Heatmap of template matching")
    
    plt.subplot(122)
    plt.imshow(fullimg_copy)
    plt.title("Detection of template")
    #Title with the method use
    
    plt.suptitle(m)
    plt.show()
    
    print('\n')
    print('\n')
    
```
